## MOVIE RECOMMENDATION SYSTEM (Content Based Recommendation)

```python
#Import necessary packages
import pandas as pd
import numpy as np
import ast

from sklearn.feature_extraction.text import CountVectorizer
```

# Data collection and processing

```python
#Loading data from CSV files into Dataframe
movies = pd.read_csv("./dataset/tmdb_5000_movies.csv")
credits = pd.read_csv("./dataset/tmdb_5000_credits.csv")
```

```python
#Merging both the datasets
movies = movies.merge(credits,on ='title')
```

```python
movies.head()
```

```python
#Keeping only the relevant columns
movies = movies[['movie_id','title','genres','overview','keywords','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

The genre column here is a string of list of dictionaries. From this dictionary we only need the name of the genre which is under the key 'name'. But for that we have to convert it to a list of dictionaries instead of a string. For that we are using the literal_eval method of the ast package.

```python
#Getting the genre names
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
        
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
#Similary getting the keyword
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

The cast column is also in the same format as genres and keywords so will apply the same method but we are only fetching the first 3 actors from the dictionary(the lead actors)

```python
#Getting the actor names from the cast column
def get_actor(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(get_actor)
movies.head()
```

```python
#Similary getting the director names from the crew column
def get_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(get_director)
movies.head()
```

Now the main objective is to get a single column by merging the crew,cast,keywords and overview columns. But the overview column is a string instead of a list, therefore converting it to a list instead

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

So here, we are eliminating the spaces between two words. Why?
Example - there is an actor named Sam Worthington and there is a director named Sam Mendes, so when we pass the text to our model it may get confused between both the Sams and may recommed wrong movies. So we convert Sam Worthington to samworthington and sam mendes to sammendes

```python
#Removing spaces between words
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#Merging all the columns to a single column named tag
movies['tags'] = movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
movies.head()
```

```python
#This is the final dataset  
new_df = movies[['movie_id','title','tags' ]]
```

```python
#Joining the list of words back to dictionary
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
#Converting everything to lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

We are using the CountVectorizer to convert the words into vectors. This will take the top 5000 words and then create a vector for each movie which signifies how many times each of those words appeared in the tags column for that movie

```python
#Initializing the CV object
cv = CountVectorizer(max_features= 5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
#Perform stemming on the tags column to get the root words
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df.head()
```

Here we are using the Cosine distance to calculate the similarity between two movies. It will calculate the angle between the two vectors and look for the vectors which have the least angular distance between them.(That means they are the most similar)

```python
#Importing packages and initializng objects
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#Function to getting the top 5 similar movies
def recommend(movie):
    try:
        movie_index = new_df[new_df['title'] == movie].index[0]
        distances = similarity[movie_index]
        movie_list = sorted(list(enumerate(distances)),reverse=True, key= lambda x:x[1])[1:6] #using enumerate to index the distance so the index is not lost after sorting
        
        for i in movie_list:
            print(new_df.iloc[i[0]].title)
        
        
    except IndexError:
        print("Movie not found")
```

```python
#Here we go!! our own movie recommendor 
recommend("Jurassic Park")
```
